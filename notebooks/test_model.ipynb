# Audio Classification Model Demo

This notebook verifies that the model is able to run, connects the data loader and the CNN correctly, and provides tensors and outputs of the correct shape.

## Imports and Paths

```python
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd().parent

if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))
```

```python
import torch

from audio_classification.data.dataset import SpectrogramDataset
from audio_classification.data.dataloader import get_data_loader
from audio_classification.models.cnn import SpectrogramCNN
```

## Load Data

```python
loader = get_data_loader(
    root_dir="../data/example_data",
    batch_size=8,
    shuffle=True
)

images, labels = next(iter(loader))

print(images.shape)
print(labels.shape)
```

## Initialize Model

```python
model = SpectrogramCNN(num_classes=6)
```

## Test a Forward Pass

```python
outputs = model(images)

print(outputs.shape)
```

## Loss Function Test

```python
criterion = torch.nn.CrossEntropyLoss()

loss = criterion(outputs, labels)

print(loss.item())
```

## Backpropagation Test

```python
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)

loss.backward()
optimizer.step()

print("backprop OK")
```
